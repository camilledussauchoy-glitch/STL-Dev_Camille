## Wavelet Operator class

```python
# Load some data
data_full = np.load(str(DATA_TEST_PATH) + "/" + "Turb_6.npy")
data = data_full[0]
print("shape data = " + str(data.shape))
print("shape data_full = " + str(data_full.shape))
plt.imshow(data)
```

```python
# Initialization of StlData for MR=False
stl_data = DataClass(array=data)
```

```python
# Construct the wavelet operator
Wave_Op = stl_data.get_wavelet_op()
```

```python
# Check the wavelet operator parameters
print("For Wave_Op:")
print("N0 = " + str(Wave_Op.N0))
print("J = " + str(Wave_Op.J))
print("L = " + str(Wave_Op.L))
print("WType = " + str(Wave_Op.WType))

print("Type wavelet_array = " + str(type(Wave_Op.wavelet_array)))
print("Shape wavelet_array = " + str(Wave_Op.wavelet_array.shape))

print("Type wavelet_array_MR = " + str(type(Wave_Op.wavelet_array_MR)))
print("dg_max = " + str(Wave_Op.dg_max))
print("j_to_dg = " + str(Wave_Op.j_to_dg))

# Fix J and L values
J = Wave_Op.J
L = Wave_Op.L
```

```python
# Check the wavelet wavelet_array_MR structure
print("Len of wavelet_array_MR = " + str(len(Wave_Op.wavelet_array_MR)))
for j in range(J):
    print("Shape wavelet_array_MR for j=" + str(j) + " is " + str(Wave_Op.wavelet_array_MR[j].shape))
```

## Plot in Fourier and real space

```python
# Plot the wavelet set in Fourier
fig, axs = plt.subplots(J, L, figsize=(2 * L, 2 * J))
axs = axs.reshape(J, L)

for j in range(J):
    for l in range(L):
        ax = axs[j, l]
        ax.imshow(Wave_Op.wavelet_array[j, l].cpu().numpy())
        ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Plot the wavelet set in Fourier, with fftshift
fig, axs = plt.subplots(J, L, figsize=(2 * L, 2 * J))
axs = axs.reshape(J, L)

for j in range(J):
    for l in range(L):
        ax = axs[j, l]
        ax.imshow(torch.fft.fftshift(Wave_Op.wavelet_array[j, l], dim=(-2,-1)).cpu().numpy())
        ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Plot the sum and quadratic sum of the wavelets

# Plot the wavelet set in Fourier, with fftshift
fig, axs = plt.subplots(1, 2, figsize=(8, 16))

axs[0].imshow(torch.fft.fftshift(Wave_Op.wavelet_array.sum(dim=(0,1)), dim=(-2,-1)).cpu().numpy())
axs[0].axis('off')
axs[1].imshow(torch.fft.fftshift((Wave_Op.wavelet_array**2).sum(dim=(0,1)), dim=(-2,-1)).cpu().numpy())
axs[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Plot the wavelet set in real space, with fftshift
fig, axs = plt.subplots(J, L, figsize=(2 * L, 2 * J))
axs = axs.reshape(J, L)

for j in range(J):
    for l in range(L):
        ax = axs[j, l]
        ax.imshow(torch.fft.fftshift(torch.fft.fft2(Wave_Op.wavelet_array[j, l]).real, dim=(-2,-1)).cpu().numpy())
        ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Plot the MR wavelet set in Fourier, with fftshift
fig, axs = plt.subplots(J, L, figsize=(2 * L, 2 * J))
axs = axs.reshape(J, L)

for j in range(J):
    for l in range(L):
        ax = axs[j, l]
        ax.imshow(torch.fft.fftshift(Wave_Op.wavelet_array_MR[j][l], 
                                     dim=(-2,-1)).cpu().numpy())
        ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Plot the MR wavelet set in real space, with fftshift
fig, axs = plt.subplots(J, L, figsize=(2 * L, 2 * J))
axs = axs.reshape(J, L)

for j in range(J):
    for l in range(L):
        ax = axs[j, l]
        ax.imshow(torch.fft.fftshift(
            torch.fft.ifft2(Wave_Op.wavelet_array_MR[j][l]).imag,
            dim=(-2,-1)).cpu().numpy())
        ax.axis('off')

plt.tight_layout()
plt.show()
```
